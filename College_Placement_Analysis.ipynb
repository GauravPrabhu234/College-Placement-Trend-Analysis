# Students who have done more than 2 projects

```python
college[college['Projects_Completed'] > 2]
```

# Students whose CGPA is greater than 9 , have less than or equal to 1 projects and got placements

```python
college[(college['CGPA']>9) & (college['Projects_Completed']<= 1) & (college['Communication_Skills']) & (college['Placement']=='Yes')]
```
